```python
# usage_example
"""
Minimal example: parcellate one fMRI file using gt_map.

Replace the values of `fmri_path` and `tr` below with your own data.
"""

from gt_map import GlasserTianParcellator

# Configuration: specify your fMRI file and its repetition time (TR)
fmri_path = "/path/to/your/fmri_file.nii.gz"  # Replace with actual file path
tr = 2.5                                      # Repetition time in seconds

# Initialize parcellator with bundled Glasser+Tian atlases (414 ROIs)
parcellator = GlasserTianParcellator()

# Parcellate the subject's fMRI data
ts = parcellator.parcellate_subject(
    fmri_path=fmri_path,
    tr=tr,
    target_tr=2.0,
    target_duration=300.0  # Standardized to 300 seconds (150 timepoints at 2s TR)
)

# Report result
if ts is not None:
    print(f"Success. Time series shape: {ts.shape}")
    print(f"Data type: {ts.dtype}")
else:
    print("Processing failed: scan duration < 300 seconds, file not found, or invalid format.")
    
```
